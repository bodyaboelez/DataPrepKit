```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder
class DataPrepKit:
    File_type =None
    File_name =None
    df = None
    def read(self):
        self.File_type = input('choose file type from (excel , csv , json) : ')
        self.File_name = input('file (name or path )is : ')
        if self.File_type == 'excel' :
            try:
                self.df = pd.read_excel(self.File_name + '.xlsx', index_col=0)
            except FileNotFoundError:
                print('File not found')
        elif self.File_type == 'csv' :
            try:
                    self.df = pd.read_csv(self.File_name + '.csv', index_col=0)
            except FileNotFoundError:
                    print('File not found')
        elif self.File_type == 'json' :
            try:
                    self.df = pd.read_json(self.File_name + '.json', index_col=0)
            except FileNotFoundError:
                    print('File not found')
        else:
                print('error : invalid input type') 
        return self.df
    def summaries(self):
        try:
            print("\nShape: {}".format(self.df.shape))
            try:
                print(self.df.head(int(input('the number of head rows is : '))))
            except ValueError :
                print('please enter vaild number')
            try:
                print(self.df.head(int(input('the number of tail rows is : '))))
            except ValueError :
                print('please enter vaild number')
            print('\ndata info : ')
            print(self.df.info())
            print('\nnubmer of missing value in \n',self.df.isnull().sum())
            print('\nUnique Values : \n',self.df.nunique())
            for category in self.df.columns:
                        if pd.api.types.is_numeric_dtype(self.df[category]):
                            print('\n'f"category : {category}")
                            print('\n'"Mean is : ", np.mean(self.df[category]))
                            print('\n'"Median is : ", np.median(self.df[category]))
                            print('\n'"Mode is : ", self.df[category].mode()[0])
                            print('\n'"Standard Deviation is : ", np.std(self.df[category]))
                            print('\n'"Minimum Value is : ", np.min(self.df[category]))
                            print('\n'"Maximum Value is : ", np.max(self.df[category]))
                        else:
                            print('\n'f"category: {category}")
                            print('\n'"Most Frequent Value : ", self.df[category].mode()[0])
                            print('\n'"number of duplication in : ", self.df[category].value_counts())
        except AttributeError:
                print('there is no data \nplease enter the data')
    def missing_data(self):
        try:
            handling_method = input('choose handling method (remove or fill) : ')
            if handling_method == 'remove' :
                self.df.dropna(inplace=True)
            elif handling_method == 'fill' :
                fill_with = input('fill with ? ')
                self.df.fillna(value=fill_with,inplace=True)
            else :
                print('please enter remove or fill')
            return self.df
        except AttributeError:
                print('there is no data \nplease enter the data') 
    def transform(self):
        try:
            df_cat=self.df.select_dtypes(np.object_)
            for col in df_cat :
                le =LabelEncoder()
                df_cat[col] = le.fit_transform(df_cat[col])
            return df_cat
        except AttributeError:
                print('there is no data \nplease enter the data')
```
